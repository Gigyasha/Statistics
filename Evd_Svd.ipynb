## Eigen Value Decomposition

Lets consider we have a matrix A which is a square symmetrix matrix having dimension N\*N , as we know the eigen vectors of square symmetrix matrix are orthogonal so matrix A contains "N" N*1 orthogonal eigen vectors and for every eigen vector,we will be having eigen value.<br>
here x1 x2.....xn are are column vectors of N*1 dimension.

\begin{bmatrix}
    \vert & \vert & \vert &\vert \\
    x_1   & x_2 & \ldots & x_n\\
    \vert & \vert & \vert & \vert
\end{bmatrix}


\begin{equation}
 \vec{AX_1} = \lambda X_1 \;\;
 \vec{AX_2} = \lambda X_2 \; ... \vec{AX_n} = \lambda X_n 
 \end{equation}

as we know,
$$ AV= \lambda V $$
A represent transformation matrix V is matrix of eigen vectors and $$ \lambda $$ represents matrix of eigen values associated with eigen vectors

begin{equation}
A.V = \begin{bmatrix}
    \vert & \vert & \vert &\vert \\
    Ax_1   & Ax_2 & \ldots & Ax_n\\
    \vert & \vert & \vert & \vert
\end{bmatrix}
 \end{equation}
 
 as we know $$ AX1= \lambda_1 X_1 \;\; and \;so \;on $$ so now the matrix of A.V will be
 \begin{equation}
 A.V = \begin{bmatrix}
    \vert & \vert & \vert &\vert \\
    \lambda_1 x_1   & \lambda_2 x_2 & \ldots & \lambda_n x_n\\
    \vert & \vert & \vert & \vert
\end{bmatrix}
 \end{equation}
 
 
 \begin{equation}
 A.V = \begin{bmatrix}
    \vert & \vert & \vert &\vert \\
    x_1   &  x_2 & \ldots &  x_n\\
    \vert & \vert & \vert & \vert
\end{bmatrix}
\begin{bmatrix}
    \lambda_1  & 0 & 0 & \ldots \\
    0   & \lambda_2 & 0 & \ldots \\
     0 & 0  & \ddots &  \vdots\\
    0   & 0 & \ldots & \lambda_n  \\
\end{bmatrix}
 \end{equation}
 
 $$ AV = \Lambda V $$
 
Multiply both side (rhs) by $$ V^{-1}$$

$$ AV V^{-1} = V \Lambda V^{-1}  \text{as we know } \;\; V V^{-1}=I \;\; \text{so the equation will be}$$ <br>
$$ A = V \Lambda V^{-1} \text{ as we know the eigen vectors of a square symmetric matrix are orthogonal and the inverse is same as transpose so we can write it as}$$ <br>
$$ A = Q \Lambda Q^{T} $$

### What EVD is actually doing?

__As we know eigen vectors inside Q are orthogonal (so prependicular and iprependicular vectors are linearly indepenedent ) so vectors inside Q are linearly independent what does that signifies is that , <br> Q contains linearly independent, orthogonal vectors it means they are also basis vectors and they spans a vector space.<br>
By doing EVD of A  matrix we have transformed the vector space spanned by vectors of A into anoother vector space spanned by the eigen vectors in Q__
